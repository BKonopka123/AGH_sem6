    funkcjonowanie algorytmu dla c1 = c2 = 2,2

Za ka≈ºdym razem nale≈ºy podaƒá ≈õrednie wyniki (warto≈õƒá funkcji przystosowania) oraz odchylenie standardowe dla 10 wywo≈Ça≈Ñ algorytmu i przedstawiƒá przyk≈Çadowe przebiegi algorytmu na wykresach (dla jednego z wywo≈Ça≈Ñ).

```python
def fun(x, y):
    return x**2 + y**2 - 20*(math.cos(math.pi * x) + math.cos(math.pi * y) - 2)

def pso(cost_func, dim=2, num_particles=30, max_iter=50, w=0.5, c1=1, c2=2):

    particles = np.random.uniform(-10, 10, (num_particles, dim))
    velocities = np.zeros((num_particles, dim))

    best_positions = np.copy(particles)
    best_fitness = np.array([cost_func(p[0], p[1]) for p in particles])
    swarm_best_position = best_positions[np.argmin(best_fitness)]
    swarm_best_fitness = np.min(best_fitness)

    for i in range(max_iter):
        r1 = np.random.uniform(0, 1, (num_particles, dim))
        r2 = np.random.uniform(0, 1, (num_particles, dim))
        velocities = w * velocities + c1 * r1 * (best_positions - particles) + c2 * r2 * (swarm_best_position - particles)

        particles += velocities

        fitness_values = np.array([cost_func(p[0], p[1]) for p in particles])

        avgs.append(np.mean(fitness_values))
        improved_indices = np.where(fitness_values < best_fitness)
        best_positions[improved_indices] = particles[improved_indices]
        best_fitness[improved_indices] = fitness_values[improved_indices]
        if np.min(fitness_values) < swarm_best_fitness:
            swarm_best_position = particles[np.argmin(fitness_values)]
            swarm_best_fitness = np.min(fitness_values)

        bests_fitness.append(swarm_best_fitness)
        if i % 10 == 0:
          print('Epoch:',i)
          print('Swarm best position:',swarm_best_position)
          print('Swarm best fitness:', swarm_best_fitness)
    deviation = math.sqrt(sum((bests_fitness-avgs[-1])**2)/len(bests_fitness))
    return swarm_best_position, swarm_best_fitness, deviation

avgs = []
bests_fitness = []
print('---------------c1=0 c2=2---------------')
solution, fitness, deviation = pso(fun, c1=0, c2=2)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=2 c2=0---------------')
solution, fitness, deviation = pso(fun, c1=2, c2=0)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=2 c2=2---------------')
solution, fitness, deviation = pso(fun, c1=2, c2=2)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=1 c2=2 w=0.1---------------')
solution, fitness, deviation = pso(fun, w=0.1)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=1 c2=2 w=0.3---------------')
solution, fitness, deviation = pso(fun, w=0.3)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=1 c2=2 w=0.5---------------')
solution, fitness, deviation = pso(fun, w=0.5)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()

avgs = []
bests_fitness = []
print('---------------c1=1 c2=2 w=0.7---------------')
solution, fitness, deviation = pso(fun, w=0.7)
print('---')
print('RozwiƒÖzanie:', solution)
print('Warto≈õƒá funkcji przystosowania:', fitness)
print('Odchylenie standardowe: ', deviation)
plt.plot(avgs)
plt.title("≈örednia warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
print('    ')
plt.plot(bests_fitness)
plt.title("Najlepsza warto≈õƒá funkcji przystosowania w ka≈ºdej iteracji [w ca≈Çej populacji]")
plt.show()
```

```python
start_time = time.time()
solution, fitness, deviation = pso(fun, w=0.5)
pso_execution_time = time.time() - start_time
print()
print("Algorytm pso:")
print("Najlepsze rozwiƒÖzanie:", solution)
print("odchylenie standardowe: ", deviation)
print("Warto≈õƒá funkcji przystosowania:", fitness)
print("Czas wykonania:", pso_execution_time)
```

```python
import numpy as np
import random
import time

def fitness_function(population):
    fitness_values = []
    for ind in population:
        x, y = ind
        fitness_values.append(fun(x, y))
    return np.array(fitness_values)

def tournament_selection(population, fitness_values, tournament_size=3):
    selected_indices = []
    while len(selected_indices) < 2:
        tournament_indices = random.sample(range(len(population)), tournament_size)
        tournament_fitness = [fitness_values[i] for i in tournament_indices]
        selected_indices.append(tournament_indices[np.argmin(tournament_fitness)])
    return selected_indices

def crossover(parent1, parent2):
    crossover_point = random.randint(0, len(parent1) - 1)
    child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
    return child1, child2

def mutate(individual, mutation_rate=0.1, mutation_scale=0.1):
    mutated_individual = individual + np.random.normal(scale=mutation_scale, size=individual.shape)
    return mutated_individual

def genetic_algorithm(pop_size=30, dim=2, max_iter=30):
    population = np.random.uniform(-10, 10, (pop_size, dim))
    for _ in range(max_iter):
        fitness_values = fitness_function(population)
        new_population = []

        while len(new_population) < pop_size:
            parent_indices = tournament_selection(population, fitness_values)
            parent1 = population[parent_indices[0]]
            parent2 = population[parent_indices[1]]

            child1, child2 = crossover(parent1, parent2)

            child1 = mutate(child1)
            child2 = mutate(child2)

            new_population.extend([child1, child2])

        population = np.array(new_population)[:pop_size]

    best_index = np.argmin(fitness_function(population))
    best_solution = population[best_index]
    best_fitness = fitness_function([best_solution])[0]

    return best_solution, best_fitness

start_time = time.time()
genetic_solution, genetic_fitness = genetic_algorithm()
genetic_execution_time = time.time() - start_time

print("Algorytm genetyczny:")
print("Najlepsze rozwiƒÖzanie:", genetic_solution)
print("Warto≈õƒá funkcji przystosowania:", genetic_fitness)
print("Czas wykonania:", genetic_execution_time)
```

Algorytm genetyczny znalaz≈Ç minimum funkcji w czasie 0.0186 sekundy, co jest minimalnie kr√≥tszym czasem ni≈º algorytm PSO. Znalezione rozwiƒÖzanie wynosi [0.01496541, 0.02259592], a warto≈õƒá funkcji przystosowania wynosi 0.07320540425620668.
Por√≥wnujƒÖc wyniki obu algorytm√≥w, mo≈ºna zauwa≈ºyƒá, ≈ºe oba znalaz≈Çy zbli≈ºone minimum funkcji przystosowania. Jednak algorytm PSO osiƒÖgnƒÖ≈Ç minimalnie ni≈ºszƒÖ warto≈õƒá funkcji przystosowania.
