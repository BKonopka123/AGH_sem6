# ZAD 1

Treść:

Proszę zaproponować swoją implementację algorytmu genetycznego w celu znalezienia maksimum funkcji f(x)=sin(100x+0.1)+x^(−x)+0.2 w przedziale [0,1], dla x=0 proszę przyjąć f(x)=0. Proszę porównać działanie algorytmu:

-Dla kodowania w naturalnym kodzie binarnym i w kodzie Graya

-Dla szansy mutacji wynoszącej 0, 0.1, 0.5 i 1.0

-Dla selekcji ruletkowej i dla selekcji progowej. W selekcji progowej dzielimy populację na dwie grupy: na γ% najlepszych i na pozostałych. Osobniki w grupie γ% najlepszych mają równą szansę na reprodukcję, pozostałe mają zerową szansę na reprodukcję. Proszę sprawdzić wyniki dla γ wynoszącego 20 i 50.

Za każdym razem proszę podać średnie wyniki dla 10 wywołań algorytmu i przedstawić przykładowe przebiegi algorytmu na wykresach (dla jednego z wywołań).


```python
class Solution:
  def __init__(self, randomize_genes = False):
    self.genes = [0] * 16
    if randomize_genes:
     for i in range(8):
      if random.random() < 0.5:
        self.genes[i] = 0
      else:
        self.genes[i] = 1

  def get_adaptation_binary(self):
    number = int("".join(str(x) for x in self.genes), 2) / 2.**16
    return math.sin(100*number + 0.1) + pow(number, -number) + 0.2

  def get_adaptation_gray(self):
    binary_code = [self.genes[0]]
    for i in range(1, len(self.genes)):
        binary_code.append(self.genes[i] ^ binary_code[i - 1])
    number = int("".join(str(x) for x in binary_code), 2) / 2.**16
    return math.sin(100*number + 0.1) + pow(number, -number) + 0.2

  def crossover(self, other_solution):
    cut_position = random.randint(0,15)
    new_solution = Solution()
    new_solution.genes[0:cut_position] = self.genes[0:cut_position]
    new_solution.genes[cut_position:] = other_solution.genes[cut_position:]
    return new_solution

  def mutation(self):
    mutation_position = random.randint(0,15)
    if self.genes[mutation_position] == 1:
      self.genes[mutation_position] = 0
    else:
      self.genes[mutation_position] = 1
```

```python
population_size = 50
iterations = 50
mutation_chances = [0, 0.1, 0.5, 1.0]
coding_types = ['binary', 'gray']
selection_types = ['roulette', 'treshold_20', 'treshold_50']


for coding_type in coding_types:

  print('---------------------CODING TYPE: ', coding_type, '---------------------\n')

  for mutation_chance in mutation_chances:

    print('---------------------MUTATION CHANCE: ', mutation_chance, '---------------------\n')

    for selection_type in selection_types:

      print('---------------------SELECTION TYPE: ', selection_type, '---------------------\n')

      population = [Solution(randomize_genes = True) for i in range(population_size)]

      best_solution = Solution()
      best_solution_adaptation = 0.
      best_iteration_found = 0

      avgs = []
      bests_local = []
      bests_global = []

      for iteration in range(iterations):
        if coding_type == 'binary':
          adaptations = [p.get_adaptation_binary() for p in population]
          local_best_solution = population[adaptations.index(max(adaptations))]
          if local_best_solution.get_adaptation_binary() > best_solution_adaptation:
            best_solution = local_best_solution
            best_solution_adaptation = local_best_solution.get_adaptation_binary()
            best_iteration_found = iteration
        else:
          adaptations = [p.get_adaptation_gray() for p in population]
          local_best_solution = population[adaptations.index(max(adaptations))]
          if local_best_solution.get_adaptation_gray() > best_solution_adaptation:
            best_solution = local_best_solution
            best_solution_adaptation = local_best_solution.get_adaptation_gray()
            best_iteration_found = iteration

        if iteration %5 == 0:
          print("Epoch: ", iteration, "; avg adaptation: ", sum(adaptations) / len(adaptations), "; best adaptation: ", max(adaptations), " best adaptation ever: ", best_solution_adaptation, " from iteration ", best_iteration_found)

        avgs.append(sum(adaptations) / len(adaptations))
        bests_local.append(max(adaptations))
        bests_global.append(best_solution_adaptation)

        if selection_type == 'roulette':
          roulette_wheel = adaptations

          for i in range(len(roulette_wheel)):
            roulette_wheel[i] -= min(adaptations)
            roulette_wheel[i] /= (max(adaptations)-min(adaptations))

          parents = [random.choices(population, weights=roulette_wheel, k = 2) for i in range(population_size)]

          children = [p[0].crossover(p[1]) for p in parents]

          for c in children:
            if random.random() < mutation_chance:
              c.mutation()
          population = children

        else:

          if selection_type == 'treshold_20':
              threshold_value = int(population_size * 0.2)
          else:
              threshold_value = int(population_size * 0.5)

          best_candidates = sorted(population, key=lambda x: x.get_adaptation_binary(), reverse=True)[:threshold_value]

          selected_parents = random.choices(best_candidates, k=population_size)

          children = [p[0].crossover(p[1]) for p in zip(selected_parents[::2], selected_parents[1::2])]

          for c in children:
              if random.random() < mutation_chance:
                  c.mutation()
          population = children


      # - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

      if coding_type == 'binary':
        adaptations = [p.get_adaptation_binary() for p in population]
        local_best_solution = population[adaptations.index(max(adaptations))]
        if local_best_solution.get_adaptation_binary() > best_solution_adaptation:
          best_solution = local_best_solution
          best_solution_adaptation = local_best_solution.get_adaptation_binary()
      else:
        adaptations = [p.get_adaptation_gray() for p in population]
        local_best_solution = population[adaptations.index(max(adaptations))]
        if local_best_solution.get_adaptation_gray() > best_solution_adaptation:
          best_solution = local_best_solution
          best_solution_adaptation = local_best_solution.get_adaptation_gray()

      print('---')
      print('Best solution: ', best_solution.genes, ' = ', int("".join(str(x) for x in best_solution.genes), 2) / 2**16., ' (correct: 0.39169)')
      print('Found in iteration: ', best_iteration_found)
      print('Largest function value found: ', best_solution_adaptation, ' (correct:  2.64358)\n')

      plt.plot(avgs)
      plt.title("Średnia wartość funkcji przystosowania w populacji")
      plt.show()
      print('    ')
      plt.plot(bests_local)
      plt.title("Przystosowanie najlepszego osobnika w każdej iteracji")
      plt.show()
      print('    ')
      plt.plot(bests_global)
      plt.title("Przystosowanie najlepszego osobnika znalezionego podczas przebiegu algorytmu")
      plt.show()
```

Wnioski:

W przypadku podziału na kod binarny i graya, kod binarny radzi sobie bardzo dobrze ze znajdowaniem maksimum, w kodzie graya pojawiąją się pewne problemy, często znajdowany jest punkt niedaleko maksimum, może to wynikać z błędnej implementacji zadania. W przypadku szansy mutacji, zauważamy że z prawdopodobieńswtem rownym 0, program nie radzi sobie ze znajdowaniem punktu, wyższe prawdopodobieństwa radzą sobie lepiej ze znajdowaniem odpowiedniego punktu. W przypadku typu selekcji, selekcaj ruletkowa radzi sobie najlepiej z powyższym problemem.

# ZAD 2

Treść:

W naszej firmie są cztery działy: A, B, C i D. Generują różne zyski i różne koszty utrzymania, a ich efektywność jest wzajemnie od siebie zależna. Maksymalna ilość pracowników w każdym dziale przy obecnych warunkach to 255 ze względu na ograniczenia budżetu i przestrzeni biurowej. Funkcja miesięcznego zysku (po uwzględnieniu kosztów) w zależności od zatrudnionych w każdym dziale pracowników ma postać:

Z(A,B,C,D)=(600*(A−20)(^2)*(B−35)^(2)) / C − (A − 50)^(2) * (D−48)^(2) + D tys. dolarów

Proszę zaproponować algorytm genetyczny (kodowanie problemu może być binarne, rzeczywistoliczbowe lub całkowitoliczbowe) i zaproponować wartość (A, B, C, D), dla której zyski firmy będą jak największe.

```python
class Solution2:
  def __init__(self):
    self.A = random.randint(1, 255)
    self.B = random.randint(1, 255)
    self.C = random.randint(1, 255)
    self.D = random.randint(1, 255)
    self.genes = [self.A, self.B, self.C, self.D]

  def get_adaptation(self):
    A = self.genes[0]
    B = self.genes[1]
    C = self.genes[2]
    D = self.genes[3]
    return (600 * (A - 20)**2 * (B - 35)**2) / C - (A - 50)**2 * (D - 48)**2 + D

  def crossover(self, other_solution):
    crossover_point = np.random.randint(1, 4)
    new_solution = Solution2()
    new_solution.genes[0:crossover_point] = self.genes[0:crossover_point]
    new_solution.genes[crossover_point:] = other_solution.genes[crossover_point:]
    return new_solution

  def mutation(self, mutation_chance):
    for i in range(4):
      if random.random() < mutation_chance:
        self.genes[i] = random.randint(1, 255)
```

```python
def selection(population):
    return random.choices(population, k=2, weights=[p.get_adaptation() for p in population])

def evolution(population, mutation_chance):
    new_population = []
    for _ in range(len(population)):
        parent1, parent2 = selection(population)
        child = parent1.crossover(parent2)
        child.mutation(mutation_chance)
        new_population.append(child)
    return new_population
```

```python
population_size = 100
iterations = 1000
mutation_chance = 0.2

population = [Solution2() for i in range(population_size)]

best_solution = Solution2()
best_solution_adaptation = 0.
best_iteration_found = 0


for iteration in range(iterations):
  adaptations = [p.get_adaptation() for p in population]

  local_best_solution = population[adaptations.index(max(adaptations))]
  if local_best_solution.get_adaptation() > best_solution_adaptation:
    best_solution = local_best_solution
    best_solution_adaptation = local_best_solution.get_adaptation()
    best_iteration_found = iteration

  if iteration % 50 == 0:
    print("Epoch: ", iteration, "; avg adaptation: ", sum(adaptations) / len(adaptations), "; best adaptation: ", max(adaptations), " best adaptation ever: ", best_solution_adaptation, " from iteration ", best_iteration_found)

  population = evolution(population, mutation_chance)

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

adaptations = [p.get_adaptation() for p in population]
local_best_solution = population[adaptations.index(max(adaptations))]
if local_best_solution.get_adaptation() > best_solution_adaptation:
  best_solution = local_best_solution
  best_solution_adaptation = local_best_solution.get_adaptation()


print('---')
print('Best solution: A:', best_solution.genes[0], ' B: ', best_solution.genes[1], ' C: ', best_solution.genes[2], ' D: ', best_solution.genes[3])
print('Found in iteration: ', best_iteration_found)
print('Cost: ', best_solution_adaptation)
```

Wnioski:

Algorytm genetyczny był skuteczną metodą do rozwiązania problemu optymalizacyjnego, polegającego na znalezieniu optymalnej liczby pracowników w każdym z czterech działów (A, B, C, D) w celu maksymalizacji zysku firmy. W trakcie działania algorytmu zauważono stopniowy wzrost wartości adaptacji (zysku firmy) w kolejnych epokach, co sugeruje, że algorytm był w stanie znajdować coraz lepsze rozwiązania w miarę upływu czasu.
